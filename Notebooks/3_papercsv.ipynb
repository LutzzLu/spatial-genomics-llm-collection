```python
import json

# Convert the Datasets_info column from JSON strings to lists of dictionaries
df['Datasets_info_parsed'] = df['Datasets_info'].apply(lambda x: json.loads(x) if pd.notna(x) and x != '' else [])

# Check the result
print("First entry as list of dictionaries:")
print(df['Datasets_info_parsed'].tolist()[0])
```

```python
# Check how many rows in df['Datasets_info'] are not empty strings and not NaN
non_empty_count = (df['Datasets_info'].notna() & (df['Datasets_info'] != '')).sum()
total_rows = len(df)
empty_count = total_rows - non_empty_count

print(f"Total rows: {total_rows}")
print(f"Non-empty Datasets_info rows: {non_empty_count}")
print(f"Empty Datasets_info rows: {empty_count}")
print(f"Percentage of non-empty rows: {(non_empty_count / total_rows) * 100:.2f}%")
```

```python
data_count = 0
available_data = 0
original_data = 0
platform_list = []
for i, _ in enumerate(df['Datasets_info_parsed']):
    if len(_) > 0:
        data_count += len(_)
        for __ in _:
            if __['available'] == True:
                # print(df['title'].tolist()[i])
                # print(__)
                # platform_list.append(__['platform'])
                available_data += 1
            if __['original_data'] == True:
                print(df['title'].tolist()[i])
                print(__)
                platform_list.append(__['platform'])
                original_data += 1
            

print(f"Total data count: {data_count}")
print(f"Available data count: {available_data}")
print(f"Original data count: {original_data}")
platform_df = pd.DataFrame(pd.Series(platform_list).value_counts()).reset_index()
platform_df.columns = ['platform', 'count']
pd.set_option('display.max_rows', None)
display(platform_df)
pd.reset_option('display.max_rows')

# print(f"Percentage of data: {(data_count / total_rows) * 100:.2f}%")


```
